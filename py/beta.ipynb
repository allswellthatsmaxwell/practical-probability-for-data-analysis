```python
from scipy.stats import beta, norm
import numpy as np, pandas as pd
import plotnine as pn

import comparer
```

```python
chicken = 40
steak = 60
chicken_steak_comparer = comparer.Comparer(
    chicken, steak, 'chicken', 'steak',
    np.arange(0.2, 0.6, 0.001))
```

```python
pn.options.dpi = 300
chicken_steak_comparer.plot
```

```python
chicken_steak_comparer.intervals()
```

```python
guinea_pig = 1
steak = 99
guinea_steak_comparer = comparer.Comparer(
    guinea_pig, steak, 'guinea pig', 'steak',
    np.arange(-0.05, 0.1, 0.001))
```

```python
guinea_steak_comparer.plot + pn.geom_vline(xintercept=0, size=2)
```

```python
guinea_steak_comparer.intervals()
```
